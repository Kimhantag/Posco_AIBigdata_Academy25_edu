```python
####  (0.6 )  (7 )   :  
####  (0.6 )    (7 )  : 
#####  (0.6 )    (7 ) : 
#####  (0.6 )    (7 ) :  

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import Normalize

# 3D  
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#    3D  
for box in data[84]:  #     
    #       
    density = box[8]
    delivery_number = box[7]
    if density > 0.6 and delivery_number > 7:
        color = 'red'  #      
    elif density > 0.6 and delivery_number <= 7:
        color = 'lightcoral'  #      
    elif density <= 0.6 and delivery_number > 7:
        color = 'lightblue'  #      
    else:
        color = 'blue'  #      

    ax.bar3d(box[3], box[4], box[5], box[0], box[1], box[2], alpha=0.8, color=color)

#    
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Box Plot with Density and Delivery Number Color Coding')

plt.show()
```
